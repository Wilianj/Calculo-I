## O problema da reta tangente

Vamos ilustrar o processo de limite através de um problema geométrico prático: o problema da reta tangente. A primeira pergunta que cabe ser feita é 'O que é uma reta tangente?'.

Para um círculo, uma reta tangente é a reta que o toca uma única vez. Na figura abaixo, qualquer outra reta que passa pelo ponto de tangência, diferente da reta tangente, cortará o círculo necessariamente em dois pontos.

```python
import numpy as np
import pylab
import matplotlib.pyplot as plt

x = np.linspace(-1.5,3,100)
y = np.linspace(-1.5,1.5,100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 1
x1 = np.linspace(-1.0,2.5,100)
y1 = (1.0-0.5**2)**0.5 - (0.5/((1.0-0.5**2)**0.5)) * (x1-0.5)

pylab.plot(0.5, (1.0-0.5**2)**0.5, 'ro')
pylab.plot(x1, y1, '-r', label='Reta tangente')
pylab.legend(loc='upper right')
plt.contour(X,Y,F,[0])
plt.show()
```

Mas e para outras curvas? Note na figura abaixo que não podemos definir uma reta tangente à curva como sendo a reta que a corta em um único ponto. Em primeiro lugar, existe mais de uma reta que passa apenas no ponto de tangência e, em segundo lugar, a reta na qual esperamos que seja a reta tangente corta a curva em mais de um ponto, quando esta se torna suficientemente distante do ponto de tangência. 

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
x2 = np.linspace(0, 8, 1000)
y2 = np.sin(np.pi*0.6) + np.cos(np.pi*0.6) * (x2-np.pi*0.6)


pylab.plot(x, y1, '-b', label='Seno')
pylab.plot(x2, y2, '-r', label='Reta tangente')
pylab.plot(np.pi*0.6,np.sin(np.pi*0.6), 'ro')
pylab.legend(loc='upper right')
pylab.ylim(-1.5, 2.0)
```

Nosso amigo Oscar já dizia que definir é limitar, mas definir gemetricamente é limitar ainda mais. Precisamos de uma definição mais algébrica para a reta tangente e, para isso, precisamos definir de forma precisa o que é a inclinação de uma reta tangente. Para tal, considere o gráfico da função $f(x)=x^2$, $x\in[-5,5]$ e uma reta tangente à curva que passa pelo ponto $(1,1)$. 

```python
x = np.linspace(-5, 5, 100)
y1 = x*x
x2 = np.linspace(-2, 5, 10)
y2 = 1 + 2 * 1 * (x2-1)


pylab.plot(x, y1, '-b', label='$y=x^2$')
pylab.plot(x2, y2, '-r', label='Reta tangente')
pylab.plot(1,1, 'ro')
pylab.legend(loc='upper right')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True, which='both')
pylab.axhline(y=0, color='k')
pylab.axvline(x=0, color='k')
pylab.ylim(-5,30)
```

A equação da reta tangente à curva $y=x^2$ que passa pelo ponto (1,1) possui a seguinte forma:

$$
y = 1+m_t(x-1),
$$
onde $m_t$ é a inclinação da reta tangente. A dificuldade surge justamente aí, já que precisamos de dois pontos para determinar
a inclinação de uma reta. Vamos, então, fazer o que sabemos, ie considerar, inicialmente, uma reta secante que passa pelo ponto de tangência $(1,1)$ e por um outro ponto da curva $(x,x^2)$, conforme ilustrado na figura abaixo. A inclinação da reta secante $(m_s)$ que passa pelos pontos $(1,1)$ e $(x,x^2)$ é definida, para qualquer $x\neq 1$, pela função

$$
m_s(x)= \frac{x^2-1}{x-1}.
$$

```python
x = np.linspace(-5, 5, 100)
y1 = x*x
x2 = np.linspace(-2, 5, 10)
y2 = 1 + 2 * 1 * (x2-1)
x3 = np.linspace(-2, 5, 10)
y3=  1 + ((16-1)/(4-1)) * (x3-1)

pylab.plot(x, y1, '-b', label='$y=x^2$')
pylab.plot(x2, y2, '-r', label='Reta tangente')
pylab.plot(x3, y3, '-g', label='Reta secante')
pylab.plot(1,1, 'ro')
pylab.plot(4,16,'go')
pylab.legend(loc='upper left')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True, which='both')
pylab.axhline(y=0, color='k')
pylab.axvline(x=0, color='k')
pylab.ylim(-5,30)
```

Notemos que conforme $x$ se aproxima de $1$ (o ponto verde se aproxima do ponto vermelho), as inclinações das retas secantes tendem à inclinação da reta tangente (ver figuras abaixo). Quanto mais próximo x estiver de $1$, mais próximo $m_s(x)$ estará de $m_t$. Podemos deixar as inclinações das retas secantes tão próximas da inclinação da reta tangente quanto quisermos, basta tomar $x$ suficientemente próximo de $1$. Isto te faz lembrar de algo?

```python
x = np.linspace(-5, 5, 100)
y1 = x*x
x2 = np.linspace(-2, 5, 10)
y2 = 1 + 2 * 1 * (x2-1)
x3 = np.linspace(-2, 5, 10)
y3=  1 + ((16-1)/(4-1)) * (x3-1)
x4 = np.linspace(-2, 5, 10)
y4=  1 + ((9-1)/(3-1)) * (x4-1)
x5 = np.linspace(-2, 5, 10)
y5=  1 + ((4-1)/(2-1)) * (x5-1)
x6 = np.linspace(-2, 5, 10)
y6=  1 + ((1.5*1.5-1)/(1.5-1)) * (x6-1)
x7 = np.linspace(-2, 5, 10)
y7=  1 + ((1.2*1.2-1)/(1.2-1)) * (x7-1)



pylab.plot(x, y1, '-b', label='$y=x^2$')
pylab.plot(x2, y2, '-r', label='Reta tangente')
pylab.plot(x3, y3, '-g', label='Reta secante')
pylab.plot(x4, y4, '-g')
pylab.plot(x5, y5, '-g')
pylab.plot(x6, y6, '-g')
pylab.plot(x7, y7, '-g')
pylab.plot(1,1, 'ro')
pylab.plot(4,16,'go')
pylab.plot(3,9,'go')
pylab.plot(2,4,'go')
pylab.plot(1.5,1.5*1.5,'go')
pylab.plot(1.2,1.2*1.2,'go')
pylab.legend(loc='upper left')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True, which='both')
pylab.axhline(y=0, color='k')
pylab.axvline(x=0, color='k')
pylab.ylim(-5,30)
```

```python
%matplotlib inline

from matplotlib import animation, rc
from IPython.display import HTML


# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()
ax.set_xlim(( -5, 5))
ax.set_ylim((-5, 30))
ax.plot(x, y1, '-b')
ax.plot(x2, y2, '-r')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.plot(1,1, 'ro')

line, = ax.plot([], [],'-g', lw=2)
line1, = ax.plot([], [], 'o-g', lw=2)

x = np.linspace(-5, 5, 100)
y1 = x*x
x2 = np.linspace(-2, 5, 100)
y2 = 1 + 2 * 1 * (x2-1)


# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    line1.set_data([], [])
    return (line,line1,)


# animation function. This is called sequentially
def animate(i):
    x3 = np.linspace(-2, 5, 100)
    y3 = 1 + ((5-0.1*i)*(5-0.1*i)-1)/((5-0.1*i)-1) * (x3-1)
    line.set_data(x3, y3)
    x4= (5-0.1*i)
    y4= (5-0.1*i)*(5-0.1*i)
    line1.set_data(x4, y4)
    return (line,line1,)


# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=40, interval=150, blit=True)

HTML(anim.to_html5_video())
```

O mesmo poderia ter sido realizado considerando $x$ à esquerda de $1$ (para $x<1$). Isto significa que o $Limite$ das inclinações das retas secantes que passam pelos pontos $(1,1)$ e $(x,x^2)$ quando $x$ tende a $1$ é igual a inclinação da reta tangente à curva $y=x^2$ que passa pelo ponto $(1,1)$. Matematicamente, podemos escrever


$$
m_t=\displaystyle \lim_{x\rightarrow 1} m_s(x)=\displaystyle \lim_{x\rightarrow 1}\frac{x^2-1}{x-1}.
$$

Este limite já foi visto na aula passada e temos que $m_t=2$. Portanto, a equação da reta tangente à curva $y=x^2$ que passa pelo ponto $(1,1)$ é dada por

$$
y = 1+2(x-1),
$$
ou seja,


$$
y = 2x-1.
$$
